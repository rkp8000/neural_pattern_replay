### Fig 1: Replay of stimulus-specific activity sequences in a network of LIF neurons
Here a stimulus is presented (~0 to 0.6 s) that drives a sequence of neuronal activations. When driven, the activated neurons excite a set of memory units, moving them to their upstates. Since the memory neurons project back to the primary neurons, the primary neurons effectively become more excitable. Therefore, when the first neuron in the sequence is later driven to spike by a "replay trigger", the downstream neurons whose activation follows will preferentially be those that were activated before. One also sees spontaneous replay of thes equences around t = 1.7s and t = 1.9s. Also, there is a slight global inhibition applied to all the primary neurons during the presentation of the stimulus.
